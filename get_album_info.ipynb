```python
import pandas as pd
```

```python
file_name = 'tracks.xlsx'
df = pd.read_excel(file_name)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df[df['Artist']=='Twista']
```

```python
url_counts = df['artwork_url'].value_counts()
unique_urls = url_counts[url_counts <= 3].index
unique_url_rows = df[df['artwork_url'].isin(unique_urls)]
df_filtered = df[~df['artwork_url'].isin(unique_urls)]
```

```python
unique_url_rows[unique_url_rows['Artist']=='Twista']
```

```python
album_df = df_filtered.groupby(['Album', 'Artist', 'Year', 'artwork_url'], sort = False)['Popularity'].mean()
```
